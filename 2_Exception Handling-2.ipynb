Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
print("The LookupError class in Python is a base class for exceptions related to lookup operations.\nIt's used when we're attempting to access an element in a collection (such as a list, dictionary, or tuple), and the specified key or index is not found or is invalid. Two common exceptions derived from LookupError are KeyError and IndexError.")
```

```python
print("KeyError")
print()

my_dict = {"apple": 3, "banana": 2, "cherry": 5}
try:
    count = my_dict["orange"]  # "orange" key doesn't exist
except KeyError as e:
    print(f"Error: {e}")
```

```python
print("IndexError")
print()

my_list = [10, 20, 30, 40, 50]
try:
    value = my_list[6]  # Index 6 is out of range
except IndexError as e:
    print(f"Error: {e}")
```

Q5. Explain ImportError. What is ModuleNotFoundError?

#### ImportError:

* ImportError is a base class for exceptions related to importing modules in Python.
* It can occur for various reasons, such as when the specified module doesn't exist, there is an issue with the module's code, or there is a problem with the import statement itself.


### ModuleNotFoundError:

* ModuleNotFoundError is a specific subclass of ImportError introduced in Python 3.6.
* It is raised when the specified module cannot be found or imported.

```python
print("ModuleNotFoundError")
print()

try:
    import non_existent_module  # Attempting to import a non-existent module
except ModuleNotFoundError as e:
    print(f"Module Not Found Error: {e}")
```
